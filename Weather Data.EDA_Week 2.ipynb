```python
Q1 = df['Dew Point Temp_C'].quantile(0.25)
Q3 = df['Dew Point Temp_C'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Dew Point Temp_C'] < lower_bound) | (df['Dew Point Temp_C'] > upper_bound)]
print(outliers)
```

```python
sns.scatterplot(x='Temp_C', y='Dew Point Temp_C', data=df)
plt.title('Temperature vs.DewPoint Temperature')
plt.xlabel('Temperature')
plt.ylabel('DewPoint Temperature')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='Temp_C', y='Wind Speed_km/h', data=df)
plt.title('Temperature vs.Wind Speed')
plt.xlabel('Temperature')
plt.ylabel('Wind Speed')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='Temp_C', y='Visibility_km', data=df)
plt.title('Temperature vs.Visibility')
plt.xlabel('Temperature')
plt.ylabel('Visibility')
plt.grid(True)
plt.show()
```

```python
new_columnname={'Date/Time':'DateTime'}
df.rename(columns=new_columnname,inplace=True)
print(df.rename)
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
df.set_index('DateTime', inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temp_C'], marker='o', linestyle='-', color='r')
plt.title('Temperature Time Series')
plt.xlabel('DateTime')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```

```python
#The above is the time series analysis of Tempreature in degrees Celcius from January 2012 to January 2013
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Wind Speed_km/h'], marker='o', linestyle='-', color='g')
plt.title('Wind Speed Time Series')
plt.xlabel('DateTime')
plt.ylabel('Wind Speed (km/h)')
plt.grid(True)
plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```

```python
#The above is my time series plot for WindSpeed.
```

```python
plt.figure(figsize=(16, 8))
plt.plot(df.index, df['Visibility_km'], marker='o', linestyle='-', color='c')
plt.title('Visibility Series')
plt.xlabel('DateTime')
plt.ylabel('Visibility (km)')
plt.grid(True)
plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.plot(df.index,df['Temp_C'],label='Temperature',color='red')
plt.plot(df.index,df['Rel Hum_%'],label='Humidity',color='blue')
plt.title('Temp&Humidity Over Time')
plt.xlabel('DateTime')
plt.ylabel('Values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#From the above chart, the variation of humidity over 1 year is much more than the variation of temperature over the same period.This would therefore make it more difficult to predict humidity in comparison to predicting temperature.
```

```python
plt.figure(figsize=(14, 8))
plt.plot(df.index,df['Temp_C'],label='Temperature',color='red')
plt.plot(df.index,df['Wind Speed_km/h'],label='WindSpeed',color='cyan')
plt.title('Temp&WindSpeed Over Time')
plt.xlabel('DateTime')
plt.ylabel('Values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Once again from the above chart, the variation of WindSpeed over 1 year is much more than the variation of temperature over the same period.This would therefore make it more difficult to predict windspeed in comparison to predicting temperature.
```

```python
plt.figure(figsize=(14, 8))
plt.plot(df.index,df['Temp_C'],label='Temperature',color='red')
plt.plot(df.index,df['Dew Point Temp_C'],label='DewPoint Temp',color='yellow')
plt.title('Temp&DewPoint Temp Over Time')
plt.xlabel('DateTime')
plt.ylabel('Values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Temperature and Dewpoint temperature had a slightly similar trend over the period of 1 year between 2012 and 2013, from the above visualization, there was not much variation in their values and so predicting both these weather features can be easy.
```
